```python
import pandas as pd
```

```python
selected_countries = ["Brazil",
                      "United States of America",
                      "Russia",
                      "Ukraine",
                      "Israel",
                      "Iran",
                      "China"]

# check if country is in column Country name in each year's dataset using pandas
for year in range(2019, 2025):
    df = pd.read_csv(f"world_happiness_{year}.csv")
    if not df['Country name'].isin(selected_countries).any():
        print(f"Warning: No selected countries found in {year} dataset.")
```

```python
for year in range(2019, 2025):
    df = pd.read_csv(f"world_happiness_{year}.csv", index_col=False)
    # df = df[df['Country name'].isin(selected_countries)]

    for col in df.columns[3:].values:
        df[col] = df[col].str.replace(",", ".")
        df[col] = df[col].apply(pd.to_numeric, errors='coerce')

    df.to_csv(f"world_happiness_{year}_pl_pre_process.csv", index=False)
```

```python
dfs = []
for year in range(2019, 2025):
    df = pd.read_csv(f"world_happiness_{year}_pl_pre_process.csv", index_col=False)
    dfs.append(df)

df_heat = pd.concat(dfs, axis=0, ignore_index=True)
df_heat.to_csv("world_happiness_pl.csv", index=False)
```

```python
normalized_country_names = {
'Czech Republic': "Czechia",
'Egypt': "Egypt, Arab Rep.",
'Hong Kong': "Hong Kong SAR, China",
'Iran': "Iran, Islamic Rep.",
'Ivory Coast': "Cote d'Ivoire",
'Kyrgyzstan': "Kyrgyz Republic",
'Laos': "Lao PDR",
'Palestine': "West Bank and Gaza",
'Republic of the Congo': "Congo, Rep.",
'Russia': "Russian Federation",
'Slovakia': "Slovak Republic",
'South Korea': "Korea, Rep.",
'The Gambia': "Gambia, The",
'Turkey': "Turkiye",
'United States of America': "United States",
'Venezuela': "Venezuela, RB",
}

df_pl["Country name"] = df_pl["Country name"].replace(normalized_country_names)
```

```python
# for each corresponding country between world_happiness and treemap, add the respective region
for i, row in df_pl.iterrows():
    country = row["Country name"]
    if country in treemap["Country Name"].values:
        region = treemap.loc[treemap["Country Name"] == country, "Region"].values[0]
        df_pl.at[i, "Region"] = region
```

```python
df_pl.to_csv("./world_happiness_pl_with_regions.csv", index=False)
```
