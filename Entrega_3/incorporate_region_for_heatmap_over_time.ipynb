```python
import pandas as pd
```

```python
normalized_country_names = {
'Czech Republic': "Czechia",
'Egypt': "Egypt, Arab Rep.",
'Hong Kong': "Hong Kong SAR, China",
'Iran': "Iran, Islamic Rep.",
'Ivory Coast': "Cote d'Ivoire",
'Kyrgyzstan': "Kyrgyz Republic",
'Laos': "Lao PDR",
'Palestine': "West Bank and Gaza",
'Republic of the Congo': "Congo, Rep.",
'Russia': "Russian Federation",
'Slovakia': "Slovak Republic",
'South Korea': "Korea, Rep.",
'The Gambia': "Gambia, The",
'Turkey': "Turkiye",
'United States of America': "United States",
'Venezuela': "Venezuela, RB",
}

treemap = pd.read_csv("./tree_map_data.csv")
```

```python
df = pd.DataFrame()

for year in range(2019, 2025):
    df_year = pd.read_csv(f"world_happiness_{year}.csv")
    df_year = df_year[["Year", "Country name", "Explained by: Social support"]]
    df_year["Country name"] = df_year["Country name"].replace(normalized_country_names)

    for i, row in df_year.iterrows():
        country = row["Country name"]
        if country in treemap["Country Name"].values:
            region = treemap.loc[treemap["Country Name"] == country, "Region"].values[0]
            df_year.at[i, "Region"] = region
    
    df_year["Explained by: Social support"] = df_year["Explained by: Social support"].str.replace(",", ".")
    df_year["Explained by: Social support"] = df_year["Explained by: Social support"].apply(pd.to_numeric, errors='coerce')

    df = pd.concat([df_year, df], axis=0)
```
