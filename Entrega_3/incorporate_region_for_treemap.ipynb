```python
import pandas as pd
```

```python
# get columns from the first row of the CSV file
# and use them as column names
treemap = pd.read_csv("./tree_map_data.csv")
world_happiness = pd.read_csv("./world_happiness_freedom.csv")
```

```python
treemap
```

```python
#order by country name
treemap = treemap.sort_values(by="Country Name")
treemap = treemap.reset_index(drop=True)
treemap
```

```python
treemap.to_csv("./tree_map_data_ordered.csv", index=False)
```

```python
world_happiness
```

```python
set(world_happiness["Country name"]) - set(treemap["Country Name"])
```

```python
normalized_country_names = {
'Czech Republic': "Czechia",
'Egypt': "Egypt, Arab Rep.",
'Hong Kong': "Hong Kong SAR, China",
'Iran': "Iran, Islamic Rep.",
'Ivory Coast': "Cote d'Ivoire",
'Kyrgyzstan': "Kyrgyz Republic",
'Laos': "Lao PDR",
'Palestine': "West Bank and Gaza",
'Republic of the Congo': "Congo, Rep.",
'Russia': "Russian Federation",
'Slovakia': "Slovak Republic",
'South Korea': "Korea, Rep.",
'The Gambia': "Gambia, The",
'Turkey': "Turkiye",
'United States of America': "United States",
'Venezuela': "Venezuela, RB",
}
```

```python
world_happiness["Country name"] = world_happiness["Country name"].replace(normalized_country_names)
```

```python
set(world_happiness["Country name"]) - set(treemap["Country Name"])
```

```python
# for each corresponding country between world_happiness and treemap, add the respective region
for i, row in world_happiness.iterrows():
    country = row["Country name"]
    if country in treemap["Country Name"].values:
        region = treemap.loc[treemap["Country Name"] == country, "Region"].values[0]
        world_happiness.at[i, "Region"] = region
```

```python
world_happiness
```

```python
# verify that all countries in world_happiness have a region
missing_regions = world_happiness[world_happiness["Region"].isnull()]
if not missing_regions.empty:
    print("Countries missing regions:")
    print(missing_regions[["Country name", "Region"]])
```
