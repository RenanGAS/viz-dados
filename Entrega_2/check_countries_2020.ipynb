```python
import pandas as pd
```

```python
df_happiness = pd.read_csv("world_happiness_2020.csv")
df_happiness
```

```python
df_regions = pd.read_csv("flourish_regions.csv")
df_regions
```

```python
list_countries_happiness = df_happiness["Country name"].values
len(list_countries_happiness)
```

```python
list_countries_flourish = df_regions["Country"].values
len(list_countries_flourish)
```

```python
list_in_happiness = list(set(list_countries_happiness) - set(list_countries_flourish)) # 17
list_in_flourish = list(set(list_countries_flourish) - set(list_countries_happiness)) # 56
```

```python
import difflib
```

```python
for word in list_in_happiness:
    similar = difflib.get_close_matches(word, list_countries_flourish, n=3)
    if len(similar) == 0:
        print(f"'{word}': '',")
    else:
        print(f"'{word}': '{similar}',")
```
