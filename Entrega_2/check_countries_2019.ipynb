```python
import pandas as pd
```

```python
df_happiness = pd.read_csv("world_happiness_2019.csv")
df_happiness.head()
```

```python
df_regions = pd.read_csv("flourish_regions.csv")
df_regions.head()
```

```python
list_countries_happiness = df_happiness["Country name"].values
len(list_countries_happiness)
```

```python
list_countries_flourish = df_regions["Country"].values
len(list_countries_flourish)
```

```python
list_in_happiness = list(set(list_countries_happiness) - set(list_countries_flourish)) # 20
list_in_flourish = list(set(list_countries_flourish) - set(list_countries_happiness)) # 55
```

```python
import difflib
```

```python
for word in list_in_flourish:
    similar = difflib.get_close_matches(word, list_countries_happiness, n=3)
    if len(similar) == 0:
        print(f"'{word}': '',")
    else:
        print(f"'{word}': '{similar}',")
```

```python
in_happiness_subs_2019 = {
    'Côte d’Ivoire': 'Ivory Coast',
    'Macedonia': 'North Macedonia',
    'DR Congo': 'Democratic Republic of the Congo',
    'South Sudan': '',
    'Russian Federation': 'Russia',
    'Türkiye': 'Turkey',
    'State of Palestine': '',
    'Czechia': 'Czech Republic',
    'Republic of Moldova': 'Moldova',
    'Republic of Korea': 'South Korea',
    'Taiwan Province of China': '',
    'North Cyprus': '',
    'Lao PDR': 'Laos',
    'Gambia': 'The Gambia',
    'Hong Kong SAR of China': '',
    'Swaziland': 'Eswatini',
    'United States': 'United States of America',
    'Congo': 'Republic of the Congo',
    'Kosovo': '',
    'Viet Nam': 'Vietnam'
}
```

```python
df_happiness_cp = df_happiness.copy()
list_country_erros = list(in_happiness_subs_2019.keys())

list_deleted = [country for country in list_country_erros if in_happiness_subs_2019[country] == '']
df_happiness_cp_cleaned = df_happiness_cp[~df_happiness_cp["Country name"].isin(list_deleted)]
df_happiness_cp_cleaned
```

```python
df_happiness_cp_cleaned["Country name"].replace(in_happiness_subs_2019, inplace=True)
```
